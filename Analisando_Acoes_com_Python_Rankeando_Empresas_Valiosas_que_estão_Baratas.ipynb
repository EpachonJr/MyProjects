# Importando Bibliotecas

```
import numpy as np
import pandas as pd
import string
import warnings
warnings.filterwarnings('ignore')
```

# Obtendo e tratando os dados

```
url = 'https://www.fundamentus.com.br/resultado.php'
df = pd.read_html(url, decimal=',', thousands='.')[0]
```

```
df
```

```
for coluna in ['Div.Yield', 'Mrg Ebit', 'Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
  df[coluna] = df[coluna].str.replace('.', '')
  df[coluna] = df[coluna].str.replace(',', '.')
  df[coluna] = df[coluna].str.rstrip('%').astype('float') / 100
```

# Analisando os dados

```
df = df[df['Liq.2meses'] > 500000]
```

```
ranking = pd.DataFrame()
ranking['pos'] = range(1,151)
ranking['EV/EBIT'] = df[ df['EV/EBIT'] > 0 ].sort_values(by=['EV/EBIT'])['Papel'][:150].values
ranking['ROIC'] = df.sort_values(by=['ROIC'], ascending=False)['Papel'][:150].values
ranking['P/VP'] = df[ df['P/VP'] > 0 ].sort_values(by=['P/VP'], ascending=True)['Papel'][:150].values
```

```
ranking
```

```
# Rankeando por EV/EBIT
```

```
a = ranking.pivot_table(columns='EV/EBIT', values='pos')
```

```
# Rankeando por ROIC
```

```
b = ranking.pivot_table(columns='ROIC', values='pos')
```

```
# Rankeando por P/B
```

```
c = ranking.pivot_table(columns='P/VP', values='pos')
```

```
# Juntando rankings
```

```
t = pd.concat([a,b,c])
t
```

```
# Retirando dados nulos e somando colocações por variável
```

```
rank = t.dropna(axis=1).sum()
rank
```

```
# Calculando Top 20
```

```
rank.sort_values()[:20]
```
